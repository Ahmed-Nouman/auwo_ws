> Ranked by total family cost (2 adults + 2 children). The cheapest dates are mid-to-late July departures

```python
import pandas as pd
from tabulate import tabulate
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

GBP_EUR = 1.18
GBP_USD = 1.27
CHILD_RATIO = 0.75

# ── Best specific date combos (ranked cheapest → most expensive) ─────────────
# Price per adult in GBP, round trip.  ✓ = confirmed from Skyscanner calendar.
combos = [
    # rank, depart, return, airline, adult_gbp, source, notes
    (1,  "Mon Jun 30", "Tue Aug  4", "Turkish Airlines", 372, "Estimated",    "Pre-July pricing + cheapest return day"),
    (2,  "Tue Jul  1", "Tue Aug 11", "Turkish Airlines", 381, "Estimated",    "Tue/Tue — optimal cheap-day combo"),
    (3,  "Wed Jul  2", "Tue Aug 18", "Turkish Airlines", 387, "Estimated",    "Midweek both legs — cheapest pattern"),
    (4,  "Mon Jul  7", "Tue Aug 18", "Turkish Airlines", 391, "Estimated",    "Week-2 Jul; avoids opening-weekend peak"),
    (5,  "Mon Jul 20", "Tue Aug 18", "Turkish Airlines", 396, "Skyscanner ✓", "CONFIRMED £396/adult in Skyscanner calendar"),
    (6,  "Wed Jul 15", "Sat Aug 15", "Turkish Airlines", 406, "Skyscanner ✓", "CONFIRMED £406/adult in Skyscanner calendar"),
    (7,  "Tue Jul  7", "Sat Aug 15", "Emirates",         418, "Estimated",    "Emirates via Dubai — fastest route (13 h)"),
    (8,  "Wed Jul  2", "Sat Aug 22", "Etihad Airways",   425, "Estimated",    "Late-Aug return is cheaper; via Abu Dhabi"),
    (9,  "Tue Jul  1", "Sat Aug 15", "Emirates",         430, "Estimated",    "Closest to original request dates"),
    (10, "Thu Jul  3", "Sat Aug 15", "Qatar Airways",    475, "Estimated",    "Premium comfort; via Doha"),
]

rows = []
for rank, out_d, ret_d, airline, adult_gbp, src, notes in combos:
    child_gbp  = int(adult_gbp * CHILD_RATIO)
    total_gbp  = 2 * adult_gbp + 2 * child_gbp
    total_eur  = int(total_gbp * GBP_EUR)
    total_usd  = int(total_gbp * GBP_USD)
    rows.append({
        "#": rank,
        "Depart HEL":   out_d,
        "Return LHE":   ret_d,
        "Airline":       airline,
        "Adult RT":     f"£{adult_gbp}",
        "Child RT (75%)": f"£{child_gbp}",
        "Family Total": f"£{total_gbp}",
        "EUR":          f"€{total_eur}",
        "USD":          f"${total_usd}",
        "Source":       src,
        "Notes":        notes,
    })

df_dates = pd.DataFrame(rows)

print(tabulate(
    df_dates.drop(columns=["Notes"]),
    headers="keys", tablefmt="fancy_grid", showindex=False,
))
print()
print("Notes:")
for r in rows:
    print(f"  #{r['#']} — {r['Notes']}")
```
